```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
#url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

url='https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_Kingdom'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify)
```

```python
table = soup.find('table', class_ = "wikitable sortable")

```

```python
print(table)
```

```python
table_columns_name = table.find_all('th')
```

```python
print(table_columns_name)
```

```python
table_columns = [title_names.text.strip() for title_names in table_columns_name]
```

```python
print(table_columns)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = table_columns)
```

```python
df
```

```python
companies_data = table.find_all('tr')
```

```python
companies_data
```

```python
for row in companies_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    # Append Dataframe by putting the data
    df.loc[len(df)] = individual_row_data
    
    
```

```python
df
```

```python
df.to_csv('Top_100_UK_Companies.csv', index=False)
```
